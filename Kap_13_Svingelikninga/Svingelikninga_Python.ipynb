```python
# EulerAndreOrden.m

"""Skript som implementerer Eulers framovermetode for ei
differensiallikning av andre orden.
Input: Startkravet y(x0)=y0, y'(x0)=u0 og F(x,y),
som definerer differensiallikninga ved
Y'=F(x,Y), der Y=[y, u]^T.
Merk at skriptet skil mellom "Y", som er 
ein vektor i R^2, og skalaren "y".
Skriptet tar talet p√• steg, N, som input.
"""

import numpy as np
import matplotlib.pyplot as plt

# Startkrav og maksimal x-verdi
x0 = 0
y0 = 2
u0 = -1
x_max = 2

def F(x, Y):
    """Differensiallikning"""
    return np.array([Y[1], -Y[1]+2*Y[0]+2*x+1])

# Steglengda
N = int(input('Gi talet p√• steg: '))
h = (x_max-x0)/N

# Lagar vektorar
x_vektor = np.linspace(x0, x_max, N)
y_vektor = np.zeros_like(x_vektor)
y_vektor[0] = y0

# Initerar x og Y
x = x0
Y = np.array([y0, u0])

# for-l√∏kke som implementerer Eulers metode
for indeks in range(0, N-1):
    # Eulers framovermetode
    Y = Y + F(x,Y)*h
    # Oppdaterer vektor med y-verdiar
    y_vektor[indeks+1] = Y[0]
    # Oppdaterer x
    x = x+h

# Plottar resultatet
plt.plot(x_vektor, y_vektor, 'r--', linewidth=2, 
         label = 'Numerisk')
plt.rcParams.update({'font.size': 15})
plt.grid(visible=True)
plt.show()
```

```python
# ForSvingelikninga.m

# Skript som implementerer Eulers midtpunktsmetode for 
# ei differensiallikning av andre orden.
# Input: Startkravet y(x0)=y0, y'(x0)=u0 og F(x,y),
# som definerer differensiallikninga ved
# Y'=F(x,Y), der Y=[y, u]^T.
# Merk at skriptet skil mellom "Y", som er ein vektor i R^2,
# og skalaren "y".
# Skriptet tar talet p√• steg N som input.

# Startkrav og maksimal x-verdi
x0 = 0
y0 = .15
u0 = 0
x_max = 50

def F(x, Y):
    """Differensiallikning"""
    return np.array([Y[1], -5*np.sin(Y[0])])

# Steglengda
N = int(input('Gi talet p√• steg: '))
h = (x_max-x0)/N

# Lagar vektorar
x_vektor = np.linspace(x0, x_max, N)
y_vektor = np.zeros_like(x_vektor)
y_vektor[0] = y0

# Initerar x og Y
x = x0
Y = np.array([y0, u0])

# for-l√∏kke som implementerer Eulers metode
for indeks in range(0, N-1):
    # Eulers midtpunktsmetode
    x_hatt = x + h/2
    y_hatt = Y + F(x,Y)*h/2
    Y = Y + F(x_hatt,y_hatt)*h
    # Oppdaterer vektor med y-verdiar
    y_vektor[indeks+1] = Y[0]
    # Oppdaterer x
    x = x+h

# Plottar resultatet
plt.plot(x_vektor, y_vektor, linewidth=1)
```
