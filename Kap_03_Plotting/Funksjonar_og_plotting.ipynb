```python
# Importerar bibliotek
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
f = np.cos(np.pi*x/20)**2 * np.sin(3*x)
plt.plot(x, f)
#plt.show()

g = np.cos(np.pi*x/20)**2
plt.plot(x, g) # feil i koden i boka - manglar "plt."
plt.show()
```

```python
x = np.linspace(0, 2, 500)
a = np.cos(2*np.pi*x)
b = np.cos(np.pi*x)
c = 2*np.cos(2*np.pi*x)
d = np.cos(np.pi*x**2)
plt.plot(x, a, 'k', linewidth=2, label='a(x)')
plt.plot(x, b, 'r--', linewidth=2, label='b(x)')
plt.plot(x, c, 'b:', linewidth=2, label='c(x)')
plt.plot(x, d, 'g-.', linewidth=2, label='d(x)')
plt.legend(loc='upper right')
plt.show()
```

```python
# Lag x og y-vektorane
x = np.linspace(0, 24, 200)
y = 3.2 * np.sin(np.pi/6*(x-3))

# Sett strrelse p teksten i figuren
plt.rcParams.update({'font.size': 15})

# Sett akseteksten
plt.xlabel('t [timar]')
plt.ylabel('T [meter]')

# Plott dei forsjellige variablane
plt.plot(x, y, linewidth=3, label='T(t)')
plt.hlines(-1, 0, 24, 'k', label='y=-1')
plt.plot(np.array([6, 18]), np.array([3.2, 3.2]), 'ro', 
         linewidth=2, label='Flo')
plt.plot(np.array([2.39, 9.61, 14.39, 21.61]), 
         -np.ones([4, 1]), 'md', linewidth=2, 
         label='T(t)=-1')

# Sett kor stort omrde ein skal plotte
plt.axis([0, 27, -4, 4])
plt.grid(visible=True)
plt.legend(loc='upper right')
plt.show()
```
