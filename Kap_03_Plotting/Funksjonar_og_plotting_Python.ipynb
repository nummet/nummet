```python
# Import
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sett størrelse på teksten i figuren
plt.rcParams.update({'font.size': 15})
```

```python
# Lag x og y-vektorane
x = np.arange(0, 24, 0.1)
y = 3.2 * np.sin(np.pi/6*(x-3))

# Sett akseteksten
plt.xlabel('t [timar]')
plt.ylabel('T [meter]')

# Plott dei forsjellige variablane.
plt.plot(x, y, linewidth=3, label="T(t)")
plt.plot(np.array([0, 24]), np.array([-1, -1]), 'k-', label="y=-1")
plt.plot(np.array([6, 18]), np.array([3.2, 3.2]), 'ro', linewidth=2, label="Flo")
plt.plot(np.array([2.39, 9.61, 14.39, 21.61]), -np.ones([4, 1]), 'gd', linewidth=2, label="T(t)=-1")

# Sett kor stort område ein skal plotte
plt.axis([0, 27, -4, 4])
plt.legend(loc="upper right")
plt.show()
```
